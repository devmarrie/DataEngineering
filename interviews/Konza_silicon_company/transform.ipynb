```python
# Read the ordes data
import pandas as pd 
orders = pd.read_csv('data/orders.csv')
orders.head()
```

```python
visits = pd.read_csv('data/visits.csv')
visits.head()
```

Transform

```python
# check for null values
orders.isnull().sum() # NA means north america which has been interprated as NaN in Super Region 
```

```python
orders.loc[orders['Country Name'] == 'US']
```

```python
orders.loc[orders['Country Name'] == 'US', 'Super Region'] = orders.loc[orders['Country Name'] == 'US', 'Super Region'].fillna('NA')
```

```python
visits.isnull().sum() # Again super region
```

```python
visits.loc[visits['Country Name'] == 'US']
```

```python
visits.loc[visits['Country Name'] == 'US', 'Super Region'] = visits.loc[visits['Country Name'] == 'US', 'Super Region'].fillna('NA')
```

```python
visits.isnull().sum()
```

```python
# Check for duplicate rows
orders.duplicated().sum()
```

```python
visits.dtypes
```

```python
# Standardize the 'Rpt Week' format in visits data
visits['Rpt Week'] = visits['Rpt Week'].str.replace('~','-')
```

```python
# Convert values written as MBL to Mobile in MObile indicator
visits.loc[visits['Mobile Indicator Name'] == 'MBL']
 
```

```python
visits['Mobile Indicator Name'] = visits['Mobile Indicator Name'].replace({'MBL': 'Mobile'})
```

```python
visits.loc[visits['Mobile Indicator Name'] == 'MBL']
```

```python
# Convert all capital letters to camel case in the Mobile Indicator column
visits['Mobile Indicator Name'] = visits['Mobile Indicator Name'].str.title()
```

```python
visits.head(130)
```

```python
orders = orders.rename(columns={'Week': 'Rpt Week'})
```

```python
orders.head()
```

```python
# Merge the two datasets based on common columns
merged_df = pd.merge(visits, orders,
                     left_on= ["Rpt Week", "Super Region", "Country Name", "Mobile Indicator Name", "Platform Type Name"],
                     right_on= ["Rpt Week", "Super Region", "Country Name", "Mobile Indicator Name", "Platform Type Name"],
                     how="inner")

merged_df.head()
```

```python
merged_df.shape[0]
```

Load

```python
# Store the new data in a csv file
merged_df.to_csv('data/merged_data.csv', index=False, 
                 columns=["Rpt Week", "Super Region", "Country Name", "Mobile Indicator Name", "Platform Type Name", "Transfer Type",  'Receive Country', 'Net Sent Amount GBP', 'Net Orders', 'Visits']) 
```
